import libraries

```python
import seaborn as sns
```

Load dataset

```python
data = sns.load_dataset('iris')
```

displying 1st 5 records

```python
data.head()
```

in the above dataset there are 5 features(4 features are numarical & 1 feature is categorical) 

```python
data = data[['petal_length','petal_width']]
```

```python
data.head()
```

```python
x = data.petal_length
x.head()
```

```python
y  = data.petal_width
y.head()
```

Plot the data points by using matplotlib lib and using scatter function

```python
from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('Iris')
```

Split the original data set into training & testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

convert the data into array by using reshape

```python
import numpy as np
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
y_train = np.array(y_train).reshape(-1,1)
y_train
```

```python
x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

To train the model we use linear regression

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

Fit the model

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
pred = model.predict(x_train)
plt.scatter(x_train,y_train,color='red',marker='+' )
plt.scatter(x_train,pred ,color='blue')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('Iris')
```

Drawing Linear Regression Line

```python
pred = model.predict(x_train)
plt.scatter(x_train,y_train,color='red',label='Datapoints',marker='+' )
plt.plot(x_train,pred ,color='blue',label='Regression line')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('Iris')
plt.legend()
```

```python
C = model.intercept_
```

```python
M = model.coef_
```
